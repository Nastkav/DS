```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
```

```python
n=10000; iter=int(n) 
```

```python
def output(S):
    mS = np.median(S)
    dS = np.var(S) 
    stdS = np.std(S)
    print(f'Mean = {mS}')
    print(f'Variation = {dS}')
    print(f'Sigma = {stdS} ')
```

```python
def uniform_distr(i):
    const = 0.5
    distr = np.random.rand(i)
    
    print(f'Uniform distribution: {distr}')
    plt.hist(distr, bins=20, facecolor="blue", alpha=0.5)
    plt.show()
    output(distr)
    
    correction = np.zeros(i)
    res = np.zeros(i)
    for k in range(i):
        correction[k] = const
        res[k] = correction[k] + distr[k]
    
    plt.plot(res)
    plt.plot(correction)
    plt.show()
    
    print('Result distribution')
    output(res)
    
    plt.hist(res, bins=20, facecolor='red', alpha=0.5, label='Result')
    plt.hist(correction, bins=20, facecolor='yellow', alpha=0.5, label='Correction')
    plt.hist(distr, bins=20, facecolor='blue', alpha=0.5, label='Normal')
    plt.show()
    
    print('Result distribution without correction')
    res_without_corr = res - correction
    output(res_without_corr)
    
    plt.hist(res, bins=20, facecolor='red', alpha=0.5, label='Result')
    plt.hist(correction, bins=20, facecolor='yellow', alpha=0.5, label='Correction')
    plt.hist(distr, bins=20, facecolor='blue', alpha=0.5, label='Normal')
    plt.hist(res_without_corr, bins=20, facecolor='green', alpha=0.5, label='Result without correction')
    plt.show()
```

```python
uniform_distr(iter)
```

```python
def normal_quadratic_distr(i):
    dm=0; dsig=5; coeff = 0.0000005
    distr1 = np.random.randn(i)
    distr = ((np.random.randn(n)) * dsig) + dm  
    
    print(f'Normal distribution: {distr}')
    plt.hist(distr, bins=20, facecolor="blue", alpha=0.5)
    plt.show()
    output(distr)

    correction = np.zeros(i)
    res = np.zeros(i)
    for k in range(i):
        correction[k] = coeff * k * k
        res[k] = correction[k] + distr[k]
    
    plt.plot(res)
    plt.plot(correction)
    plt.show()
    
    print('Result distribution')
    output(res)
    
    plt.hist(res, bins=20, facecolor='red', alpha=0.5, label='Result')
    plt.hist(correction, bins=20, facecolor='yellow', alpha=0.5, label='Correction')
    plt.hist(distr, bins=20, facecolor='blue', alpha=0.5, label='Normal')
    plt.show()
    
    print('Result distribution without correction')
    res_without_corr = res - correction
    output(res_without_corr)
    
    plt.hist(res, bins=20, facecolor='red', alpha=0.5, label='Result')
    plt.hist(correction, bins=20, facecolor='yellow', alpha=0.5, label='Correction')
    plt.hist(distr, bins=20, facecolor='blue', alpha=0.5, label='Normal')
    plt.hist(res_without_corr, bins=20, facecolor='green', alpha=0.5, label='Result without correction')
    plt.show()
    
    
```

```python
normal_quadratic_distr(iter)
```
